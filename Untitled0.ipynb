```
X=np.random.random((7254,200,200,3))
```

```
y = np.array(train.drop(['Id', 'Genre'],axis=1))
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=1, test_size=0.2)
```

```
def model_exp(opt,lr,act,d,hyp):
  model = Sequential()
  model.add(Conv2D(filters=16, kernel_size=(5, 5), activation=act, input_shape=(200,200,3)))
  model.add(MaxPooling2D(pool_size=(2, 2)))
  model.add(Dropout(d))
  model.add(Conv2D(filters=32, kernel_size=(5, 5), activation=act))
  model.add(MaxPooling2D(pool_size=(2, 2)))
  model.add(Dropout(d))
  model.add(Conv2D(filters=64, kernel_size=(5, 5), activation=act))
  model.add(MaxPooling2D(pool_size=(2, 2)))
  model.add(Dropout(d))
  model.add(Conv2D(filters=64, kernel_size=(5, 5), activation=act))
  model.add(MaxPooling2D(pool_size=(2, 2)))
  model.add(Dropout(d))
  model.add(Flatten())
  model.add(Dense(128, activation=act))
  model.add(Dropout(d))
  model.add(Dense(64,  activation=act))
  model.add(Dropout(d))
  model.add(Dense(25, activation='sigmoid'))
  if opt:
    opt = Adam(lr=lr,beta_1=0.9,beta_2=0.999,epsilon=1e-07,amsgrad=hyp)
    model.compile(loss='binary_crossentropy', optimizer=opt,metrics=['binary_accuracy','accuracy','categorical_accuracy'])
    test=model.fit(X_train, y_train, epochs=10, validation_data=(X_test, y_test), batch_size=500,verbose=0)
    return test
  else:
    opt = SGD(lr=lr,decay=1e-6, momentum=0.99,nesterov=hyp)
    model.compile(loss='binary_crossentropy', optimizer=opt, metrics=['binary_accuracy','accuracy','categorical_accuracy'])
    test=model.fit(X_train, y_train, epochs=10, validation_data=(X_test, y_test), batch_size=500,verbose=0)
    return test
```

```
lr=np.array([0.1,0.01,0.001,0.000001])
act=["relu","sigmoid","tanh",]
drp=np.array([0.2,0.3,0.4,0.5])
res=np.zeros((192,8,10))
for i in range(2):
  for j in (lr):
    for k in act:
      for l in drp:
        for m in range(2):
          print("optim: ",i,"   lr:",j,"  activation: ",k,"  dropout: ",l,"   _________: ",m)
          val=model_exp(i,j,k,l,m)
          res[i]=np.array((val.history['accuracy'],val.history['val_accuracy'],val.history['loss'],val.history['val_loss'],val.history['binary_accuracy'],val.history['val_binary_accuracy'],val.history['categorical_accuracy'],val.history['val_categorical_accuracy']))
          #print(np.max(val.history['accuracy']),np.max(val.history['val_accuracy']),np.min(val.history['loss']),np.min(val.history['val_loss']),np.max(val.history['binary_accuracy']),np.max(val.history['val_binary_accuracy']),np.max(val.history['categorical_accuracy']),np.max(val.history['val_categorical_accuracy']))
```

```
def CNN(filt_size,k_size,X_train, X_test, y_train, y_test):
  model = Sequential()
  model.add(Conv2D(filters=4*1,  kernel_size=(4,4), activation="relu", input_shape=(X_train.shape[1],X_train.shape[2],3)))
  model.add(Conv2D(filters=4*2,  kernel_size=(4,4), activation="relu",padding='valid',use_bias='True'))
  model.add(Conv2D(filters=4*4,  kernel_size=(4,4), activation='relu',padding='valid',use_bias='True'))
  model.add(Conv2D(filters=4*8,  kernel_size=(4,4), activation='relu',padding='valid',use_bias='True'))
  model.add(Conv2D(filters=4*16,  kernel_size=(4,4), activation='relu',padding='valid',use_bias='True'))
  #model.add(AveragePooling2D(pool_size=4))
  model.add(Conv2D(filters=4*32, kernel_size=(4,4), activation='relu',padding='valid',use_bias='True'))
  model.add(Conv2D(filters=4*64, kernel_size=(4,4), activation='relu',padding='valid',use_bias='True'))
  model.add(AveragePooling2D(pool_size=4))
  model.add(Conv2D(filters=4*16, kernel_size=(4,4), activation='relu',padding='valid',use_bias='True'))
  model.add(Conv2D(filters=4*8,  kernel_size=(4,4), activation='relu',padding='valid',use_bias='True'))    
  model.add(AveragePooling2D(pool_size=4))
  model.add(Conv2D(filters=4*4,  kernel_size=(4,4), activation='relu',padding='valid',use_bias='True'))
  model.add(Conv2D(filters=4*2,  kernel_size=(4,4), activation="relu",padding='valid',use_bias='True'))
  model.add(Flatten())
  model.add(Dense(8, activation='relu'))
  model.add(Dense(25, activation='sigmoid'))
  sgd = SGD(lr=1e-3, decay=1e-6, momentum=0.9, nesterov=True)
  model.compile(loss='binary_crossentropy', optimizer=sgd, metrics=['accuracy'])
  model.fit(X_train, y_train, epochs=10, validation_data=(X_test, y_test), batch_size=50,shuffle='True',use_multiprocessing='True',verbose=0)
  _, accuracy=model.evaluate(X_test, y_test, batch_size=100, verbose=0)
  return accuracy
```

```
def summarize_results(scores, params):
  print(scores, params)
  for i in range(len(scores)):
   m, s = np.mean(scores[i]), np.std(scores[i])
   print('Param=%d: %.3f%% (+/-%.3f)' % (params[i], m, s))
  plt.figure(figsize=(20,10))
  plt.boxplot(scores, labels=params)
```
